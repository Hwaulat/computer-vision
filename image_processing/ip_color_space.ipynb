```python
import matplotlib.pyplot as plt
import numpy as np
import cv2
```

Function to display images in Jupyter Notebooks and Google Colab


```python
def plt_imshow(title, jeruk):
    if len(jeruk.shape) == 3:
        jeruk = cv2.cvtColor(jeruk, cv2.COLOR_BGR2RGB)
        plt.imshow(jeruk)
        plt.title(title)
        plt.grid(False)
        plt.show()
```

```python
# since we are using Jupyter Notebooks we can replace our argument
# parsing code with *hard coded* arguments and values
args = {
    "image": "batik1.jpg",
}
```

```python
# load the original image and show it
jeruk = cv2.imread('jeruknipis.png')
plt_imshow("RGB", jeruk)

for (name, chan) in zip(("Red", "Green", "Blue"), cv2.split(jeruk)):
    cv2.imshow(name, chan)
```
